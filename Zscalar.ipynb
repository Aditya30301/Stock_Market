```python
import requests
from bs4 import BeautifulSoup
import csv 
import time
%matplotlib inline
import matplotlib.pyplot as plt



def get_val(Index):
    index=Index
    url="https://finance.yahoo.com/quote/ZS/"
    source_code=requests.get(url)
    plain_text=source_code.text
    soup=BeautifulSoup(plain_text)
    for stockval in soup.findAll('span',{'class':'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)'}):
        value=stockval.text
        print(value)
        data_csv(value,Index)
        
        
def data_csv(value,Index):
    t = time.localtime()
    current_time = time.strftime("%H:%M:%S", t)
    current_time=[current_time]
    file_name="Zscalar_Stock.csv"
    value=[Index]+current_time+[value]
    
    
    with open(file_name,'a') as csvfile:
        logger=csv.writer(csvfile, lineterminator = '\n')
        logger.writerow(value)

def plot():
    x=[]
    y=[]
    with open('Zscalar_Stock.csv', 'r') as csvfile:
        plots= csv.reader(csvfile)
        for row in plots:
            x.append((row[1]))
            y.append((row[2]))
    plt.plot(x,y, marker='o')
    plt.title('Data from the CSV File: People and Expenses')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()
        
        
        
        
def main():
    counter=1
    while True:
        get_val(counter)
        plot()
        counter=counter+1
        time.sleep(2)
        
if __name__=="__main__":
    
    file_name="Zscalar_Stock.csv"
    fields=[0,0,0]
    '''with open(file_name,'w+') as csvfile:
        csv_writer=csv.writer(csvfile,lineterminator = '\n')
        csv_writer.writerow(fields)'''
    main()



    
```
